````python
from math import sqrt

#请在下面编写你的类

class Point:
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y
    def Draw(self):
        print("(%f,%f)" % (self.x,self.y))
    def distance(self,x1,y1):
        d=sqrt((self.x-x1)**2+(self.y-y1)**2)
        print(d)
    def __add__(self,x1):
        x=x1.x+self.x
        y=x1.y+self.y
        z=Point(x,y)
        return z
    def __sub__(self,x1):
        x=self.x-x1.x
        y=self.y-x1.y
        z=Point(x,y)
        return z
    
if __name__ == '__main__':

    x1=Point(1,2)
    x1.Draw()

    x2=Point(3,4)
    x2.Draw()

    z1=x1+x2
    z1.Draw()

    z2=x1-x2
    z2.Draw()




````

## 3.在图像目标检测中，通常用一个轴向包围盒表示图像中的某个目标的位置和大小，轴向包围盒是一个矩形，可以由左上角与右下角坐标来唯一确定。除了坐标之外，目标检测程序通常会给每一个包围盒打一个分值，分值是一个正数。请设计一个类Box表示轴向包围盒。

要求：

1） 在构造函数中指定包围盒的左上角坐标和右下角坐标，可以使用前面定义的Point类。

2）设计一个类方法，用于计算两个包围盒的交并比IoU,计算方法如下：

假设包围盒矩形A,B的坐标分别为$A=(x_{1}^A,y_1^A,x_2^A,y_2^A),B=(x_{1}^B,y_1^B,x_2^B,y_2^B)$,其中$(x_1,y_1)$是左上角坐标，$(x_2,y_2)$是右下角坐标。他们的交并比IoU(A,B)表示A，B重叠部分($A\cap B$)的面积与他们的联合部分($A\cup B$)总面积之比。可以按照以下方式计算：

$x_1^I = max(x_1^A,x_1^B); y_1^I = max(y_1^A,y_1^B); x_2^I = min(x_2^A,x_2^B); y_2^I = min(y_2^A,y_2^B)$

交集面积为$I = (x_2^I - x_1^I)\times (y_2^I - y_1^I)$

$IoU(A,B)=\frac{I}{A+B-I}$，其中A,B分别表示矩形A,B的面积。

3）重载>运算符，根据包围盒的分值大小对包围盒进行大小比较。比如：A的分值为0.9，B的分值为0.7，则A>B为True。

4）设计一个方法display，输出包围盒的坐标以及分值。

````python
#请在下面编写类代码
class Box(Point):
    def __init__(self,x,y,x2,y2,score):
        Point.__init__(self,x,y)
        self.x2=x2
        self.y2=y2
        self.P1=Point(x,y)
        self.P2=Point(x2,y2)
        self.score=score

    def display(self):
        print('左上角坐标:',self.P1)
        print('右下角坐标:',self.P2)
        print('The score is %.2f'%self.score)

    def __Compare__(self,other):
        if self.score>other.score:
            return True
            
    def Iou(cls,A,B):
        if A.x2 < B.x:
            Iou = 0
        else:
            x1=max(A.x,B.x)
            y1=min(A.y,B.y)
            x2=min(A.x2,B.x2)
            y2=max(A.y2,B.y2)
            SI=abs((x2-x1)*(y2-y1))
            SA=(A.x2-A.x)*(A.y-A.y2)
            SB=(B.x2-B.x)*(B.y-B.y2)
            Iou=SI/(SA+SB-SI)
        return Iou
````

在图像目标检测中，非极大值抑制（Non-Maximum Suppression，NMS）是一个通用的筛选包围盒的方法。该算法从相互重叠的包围盒中，找出得分最高的，并删除其它与之重叠度超过给定阈值的包围盒。
编写代码，利用上面定义的包围盒类，实现NMS算法。

NMS算法如下：

首先，对所有包围盒按照分值从大到小排序，得到排序后的包围盒序列为$\{B_i\}_{i=1}^{n}$。

````python
#在下面编写函数，实现NMS算法
from random import randint
def Create_Box(n):
    B=[]
    for i in range(1,n+1):
        a=randint(1,100)
        b=randint(1,100)
        c=randint(1,100)
        d=randint(1,100)
        box=Box(a,b,c,d)
        B.append(box)
    return B

def SelectSort(B):
    for i in range(len(B)):
        k=i
        for j in range(i+1,len(B)):
            if B[j]<B[k]:
                k=j
        if k!=i:
            t=B[i]
            B[i]=B[k]
            B[k]=t
    return B

def NMS(B,t):
    suppressed=[]
    for i in range(len(B)):
        for j in range(i+1,len(B)):
            if j in suppressed:
                continue
            if B[i].Iou(B[i],B[j])>t:
                suppressed.append(j)
    results=[B[i] for i in range(len(B)) if i not in suppressed]
    return results


````
